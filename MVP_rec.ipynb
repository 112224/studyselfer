```python
def train(idx):
    train_x = student_trys[student_trys['student_id']==idx].loc[:,['question_id','part_L','part_M','part_S','difficulty','right']]
    return train_x
```

## hyper parameters

```python
epoch = 30
input_size = 7
batch_size = 400
dropout = 0.3
num_of_classes = 5
```

```python
사용자의 한 회분 이력 -> 모델 -> 등급 예측값

한 문제에 대한 이력 -> 모델 -> 등급예측값
3번문제 4번 -> 맞춘 사람들 -> 등급이 1~5 다양(5등급)   -> [0,0,0,0,0,1]






model = keras.models.Sequential()
model.add(layers.GRU(256,input_shape = (None,6)))
model.add(layers.Dense(5,activation = 'softmax'))
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam')
model.summary()
```
